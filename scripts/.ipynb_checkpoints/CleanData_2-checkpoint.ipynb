```python
import pandas as pd

data = pd.read_csv('../data/GSE12345_samples.csv', index_col = 0)

print("First few rows of the data:")
data.head()
#data.columns
#data.dtypes
```

```python
# Check for duplicates
duplicates_num = data.drop_duplicates().sum()
print(f"\nNumber of duplicate rows:\n{duplicates_num}")
```

```python
# check for missing values

missing_val_num = data.isnull().sum().sum()
print(f"\fNumber of missing values: {missing_val_num}")
```

```python
# Check for basic statistics

summary_stats = data.describe()
print(f"\nSummary statistics before normalization:\n{summary_stats}")
```
